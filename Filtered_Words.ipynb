```python
import json
from textwrap import dedent as d
import pandas as pd
import plotly.graph_objects as go
import numpy as np
import dash
from dash import dcc
from dash import html
import plotly.express as px
from dash.dependencies import Input, Output
from jupyter_dash import JupyterDash
import warnings

warnings.simplefilter(action='ignore', category=FutureWarning)

# app info
app = JupyterDash(__name__)
styles = {
    'pre': {
        'border': 'thin lightgrey solid',
        'overflowX': 'scroll'
    }
}

# data

df = px.data.gapminder().query("continent=='Oceania'")

app.layout = html.Div([
    html.Div(
        html.Div([
            dcc.Dropdown(
               options=[
                   {'label': 'COVID', 'value': 'COVID'},
                   {'label': 'Coronavirus', 'value': 'Coronavirus'},
                   {'label': 'Pandemia', 'value': 'Pandemia'},
                   {'label': 'Vacuna', 'value': 'Vacuna'},
                   {'label': 'Cubrebocas', 'value': 'Cubrebocas'},
                   {'label': 'AMLO', 'value': 'AMLO'},
                   {'label': 'Pfizer', 'value': 'Pfizer'},
                   {'label': 'AstraZeneca', 'value': 'AstraZeneca'},
               ], placeholder="Seleccione un termino para filtrado", id='comboterms', style={"width": "60%"}),

            html.Pre(id='txttweet', style=styles['pre']),
    ]), )
])

# container for clicked points in callbacks
store = []
```

```python
@app.callback(Output('txttweet', 'children'), Input('comboterms', 'value'))
def update_text(comboterms_name):
    if comboterms_name is not None:
        return html.Div([
                    html.Pre('Valor Seleccionado ' + comboterms_name)
                ])

app.run_server(mode='external', port = 7077, dev_tools_ui=True, dev_tools_hot_reload =True, threaded=True)
```
