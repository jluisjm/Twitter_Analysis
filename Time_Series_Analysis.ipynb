```python
import warnings, json, numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.stattools import pacf

ruta = 'C://TweetsPositivosMuertes.csv'
df = pd.read_csv(ruta, header=0)
plot_pacf(df['Polarity'], title='PACF: Modo Normal')
```

```python
plot_pacf(df['Polarity'], title='PACF: Plot lag 14', lags=14)
```

```python
plot_pacf(df['Polarity'], title='PACF: Plot lag 200', lags=200)
```
