<h2>En este notebook se inicia el trabajo para explorar diversas características subyacentes en los tweets</h2>

```python
import seaborn as sns
import matplotlib.pyplot as plt
import spacy, string, re, sklearn, pandas as pd, numpy as np
ruta = 'Tweet4LDA.csv'
import matplotlib.pyplot as plt
%matplotlib inline
from subprocess import check_output
from wordcloud import WordCloud
```

<h3>Se aplica una función para limpiar el texto, como se describe a continuación:</h3>

```python
def clean_text(text):
#convertir a minusculas   
  text = text.lower()
#quitar texto entre corchetes
  text = re.sub(r'\[.*?\]', '', text)
#quitar puntuacion 
  text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text) 
#eliminar palabras que contengan números
  text = re.sub(r'\w*\d\w*', '', text)
  return text

df = pd.read_csv(ruta, header=None)
df["tweets_df_clean"] = pd.DataFrame(df.Tweet.apply(lambda x: clean_text(x)))
```

<h3>Al texto que se ha limpiado, se aplica una función de lematización, es decir, se extrae su raíz sintáctica para poder reducir el tamaño del Corpus que se estará analizando<BR>
Por lo que palabras como: vacunar, vacunación, vacuné, vacunas, se reducirían a una raíz como: 'vacun' </h3>

```python
nlp = spacy.load("es_core_news_md")
def lemmatizer(text):        
    sent = []
    doc = nlp(text)
    for word in doc:
        sent.append(word.lemma_)
    return " ".join(sent)

df["tweets_lemma"] = pd.DataFrame(df.tweets_df_clean.apply(lambda x: lemmatizer(x)))
df.head(20)
```
